```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('venice.PNG')

# Plottin without fixing the BRG stuff on matplotlib
plt.figure(figsize=(12, 8))
plt.imshow(img)
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,8))
plt.imshow(rgb_img)
```

```python
img_copy = rgb_img.copy()
```

```python
cv2.rectangle(img_copy, (820, 620), (1300, 950), (255, 0, 0), 5)
cv2.putText(img_copy, 'Gondola', (820, 620), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 4, (255, 0, 0), 2)

cv2.rectangle(img_copy, (580, 180), (1500, 780), (0, 255, 0), 5)
cv2.putText(img_copy, 'Building', (580, 170), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 4, (0, 255, 0), 2)

plt.figure(figsize=(12, 8))
plt.imshow(img_copy)
```
