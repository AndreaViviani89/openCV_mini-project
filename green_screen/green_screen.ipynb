```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
man_green_screen = cv2.imread('3.jpg')
background = cv2.imread('back.jpg')
```

```python
type(man_green_screen)
```

```python
mgs_hsv = cv2.cvtColor(man_green_screen, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,8))
plt.imshow(mgs_hsv)
```

```python
back_hsv = cv2.cvtColor(background, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,8))
plt.imshow(back_hsv)
```

```python
img = man_green_screen.copy()

img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

u_green = np.array([100, 255, 100]) 
l_green = np.array([0, 100, 0])

mask = cv2.inRange(img, l_green, u_green)

masked_img = img.copy()
masked_back = background.copy()

masked_img[mask != 0] = [0,0,0]
masked_back[mask == 0] = [0,0,0]

merged = masked_img + masked_back
merged = cv2.cvtColor(merged, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,20))
plt.imshow(merged)
```
